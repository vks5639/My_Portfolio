# Downloading the Sequences

Download the actual nucleotide sequence for each of the selected COVID RNA sequences. Use a Python library called Biopython that allows us, among many other things, to look up RNA and DNA sequences by their accession number and then download them.

Import Entrez from Bio.

Apply this function to each of the sequences identified earlier to create a sequence data dictionary.

```python
from Bio import Entrez
Entrez.email = "vikashsingh5639@gmail.com"
```

```python
def download_sequence(id_code):
    handle = Entrez.esearch(db = "nucleotide", term=id_code, retmax="1")
    record = Entrez.read(handle)
    handle = Entrez.efetch(db = "nucleotide", id=record["IdList"][0], rettype="fasta", retmode="text")
    return handle.read()
```

```python
sequence_data = {}
for sequence in sequences:
    sequence_data[sequence] = {"fasta": download_sequence(sequence)}
```

```python
sequence_data
```

# Aligning RNA Sequences

Now that we have obtained the COVID nucleotide sequences in a readable format, our next step is to align them based on the number and order of nucleotides. This alignment process will help us identify how the virus has mutated over time.

To provide context, mutations in RNA sequences typically manifest in three ways:

1. Deletion: a nucleotide is missing in the variant sequence compared to the original sequence.
2. Substitution: a nucleotide from the original sequence is replaced by a different nucleotide in the variant sequence.
3. Insertion: an additional nucleotide appears in the variant sequence compared to the original sequence.

These mutations illustrate how an RNA strand, or in our case, the initial COVID virus, can evolve into different variants.

Our objective is to determine which segments of two chosen RNA sequences align with each other and which segments do not. Essentially, we aim to identify the differences between the sequences. To achieve this, we will utilize an alignment algorithm from Biopython. By calculating the alignment score between the sequences, we can draw conclusions about their differences.

Create a pairwise aligner class, and apply an alignment algorithm to it.

Find the alignment score between the reference sequence and any other sequence from the dictionary.

Analyze the result about the degree of the virus's mutation.

```python
from Bio import Align
aligner = Align.PairwiseAligner()
```

```python
aligner.algorithm
```

```python
score = aligner.score(sequence_data["NC_045512.2"]["parsed"], sequence_data["OM061695.1"]["parsed"].seq)
```

# Checking Alignment Across Variants

We can look at this score between all of the sequences from our dictionary. In other words, we need to compare the sequences pairwise in terms of their alignment scores and extract insights from those numbers.

Create an empty comparison matrix to further compare the sequences against each other and fill the values of the alignment scores into the matrix.

```python
import numpy as np
comparisons = np.zeros((4,4))
```

# Color Coding Mismatches

Two issues with the result from the previous screen:

1. It doesn't show where exactly the mismatch happened.

2. It doesn't show the type of mismatch (insertion, deletion, or substitution).

Fix it is by applying color coding in the HTML format using IPython. The idea is to display in different colors the nucleotide composition of the following:

All the portions of both sequences that are aligned

All the types of mismatches, including insertions, deletions, and substitutions

```python
from IPython.display import HTML
```

```python
def color_print(s, color="black"):
    return "<span style='color:{}'>{}</span>".format(color,s)
```

```python
seq1_end = None
seq2_end = None
display_seq = []
for alignments in zip(delta_alignment.aligned[0], delta_alignment.aligned[1]):
    
    if seq1_end and seq2_end:
        seq1_mismatch = seq1[seq1_end:alignments[0][0]]
        seq2_mismatch = seq2[seq2_end:alignments[1][0]]
        if len(seq2_mismatch)==0:
            display_seq.append(color_print(seq1[seq1_end:alignments[0][0]], "red"))
        elif len(seq1_mismatch)==0:
            display_seq.append(color_print(seq2[seq2_end:alignments[1][0]], "green"))
        else:
            display_seq.append(color_print(seq2[seq2_end:alignments[1][0]], "blue"))
    
    display_seq.append(seq1[alignments[0][0]:alignments[0][1]])
    
    seq1_end = alignments[0][1]
    seq2_end = alignments[1][1]
```

```python
display_seq = [str(i) for i in display_seq]
```

```python
display(HTML('<br>'.join(display_seq)))
```
